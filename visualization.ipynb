# LABELS (etiketler)

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x,y)
```

```python
# gradiğe başlık verdik.
plt.title("Bu ana başlık")

# x ekseni isimlendirmesi
plt.xlabel("x ekseni isimlendirmesi")

# y ekseni isimlendirmesi
plt.ylabel("y ekseni isimlendirmesi")

# grafğie okunabilirlik açısından ızgara ekliyoruz
plt.grid()

plt.plot(x,y)
    
```

# SUBPLOTS (etiketler)
**birden fazla görselin gösterilmesi çalıştırılması**

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.subplot(1,2,1)  
# 1 satırlık 2 sütunluk bir grafik oluşturuyorum ve bunun şu anda 1. grafiiğini oluşturuyorum

plt.title("1.grafik")
plt.plot(x,y)

# 2.sinide çizdiriyoruz. 

x = np.array([8, 8, 9, 9, 10, 15, 11, 12, 15])
y = np.array([24, 20, 26, 27, 280, 29, 30, 30, 30])
plt.subplot(1,2,2)  
plt.title("2.grafik")
plt.plot(x,y)
```

```python
# 3 grafiği 1 satır 3 sütun olarak konumlamak.
#plot 1
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(1,3,1)  
plt.title("1.grafik")
plt.plot(x,y)
 
#plot 2
x = np.array([8, 8, 9, 9, 10, 15, 11, 12, 15])
y = np.array([24, 20, 26, 27, 280, 29, 30, 30, 30])
plt.subplot(1,3,2)  
plt.title("2.grafik")
plt.plot(x,y)

#plot 3
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(1,3,3)  
plt.title("3.grafik")
plt.plot(x,y)
```

# SEABORN İLE VERİ GÖRSELLEŞTİRME

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("/kaggle/input/seaborn-tips-dataset/tips.csv")
df.head()
```

```python
df["sex"].value_counts()
```

# kategorik değişkenleri görselleştirmek

```python
# şimdi dağılmış olan frekansların görselleştirmesini yapıyoruz..
sns.countplot(x=df["sex"],data=df)
plt.show()
```

```python
#aynı grafik matplotlibde nasıl gerçekleşirdi.
df["sex"].value_counts().plot(kind="bar")
plt.show()
```

# sayısal değişkenleri görselleştirmek

```python
sns.boxplot(x=df["total_bill"])
plt.show()
```

```python
#pandas içerisinde yer alan histogram- kullanımı
df["total_bill"].hist()
plt.show()
```
