# SUBPLOTS (etiketler)
**birden fazla gÃ¶rselin gÃ¶sterilmesi Ã§alÄ±ÅŸtÄ±rÄ±lmasÄ±**

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.subplot(1,2,1)  
# 1 satÄ±rlÄ±k 2 sÃ¼tunluk bir grafik oluÅŸturuyorum ve bunun ÅŸu anda 1. grafiiÄŸini oluÅŸturuyorum

plt.title("1.grafik")
plt.plot(x,y)

# 2.sinide Ã§izdiriyoruz. 

x = np.array([8, 8, 9, 9, 10, 15, 11, 12, 15])
y = np.array([24, 20, 26, 27, 280, 29, 30, 30, 30])
plt.subplot(1,2,2)  
plt.title("2.grafik")
plt.plot(x,y)
```

```python
# 3 grafiÄŸi 1 satÄ±r 3 sÃ¼tun olarak konumlamak.
#plot 1
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(1,3,1)  
plt.title("1.grafik")
plt.plot(x,y)
 
#plot 2
x = np.array([8, 8, 9, 9, 10, 15, 11, 12, 15])
y = np.array([24, 20, 26, 27, 280, 29, 30, 30, 30])
plt.subplot(1,3,2)  
plt.title("2.grafik")
plt.plot(x,y)

#plot 3
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(1,3,3)  
plt.title("3.grafik")
plt.plot(x,y)
```
